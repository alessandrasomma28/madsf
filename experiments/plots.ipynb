```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Generate plots

```python
df = pd.read_csv("/Users/beyzaeken/Desktop/sfdigitalmirror/sumoenv/scenarios/underground_alarm/social_groups/21111008_21111014/sf_final_metrics.csv")
# Apply rolling mean and drop NaNs once
df = df.rolling(window=300).mean().dropna()

# Plot each column in a separate graph
for column in df.columns:
    plt.figure()  # Create a new figure for each plot
    plt.plot(df[column], label=column)
    plt.title(f"Rolling Average of {column}")
    plt.xlabel("Index")
    plt.ylabel(column)
    plt.legend()
    plt.grid(True)
    plt.show()
```
